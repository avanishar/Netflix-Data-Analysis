## Problem Statement
The objective of this project is to analyze Netflix content data to understand
trends in movies and TV shows, popular genres, content growth over the years,
ratings distribution, and country-wise production. This analysis helps in
understanding Netflix’s content strategy.


```python
import pandas as pd
import numpy as np
df = pd.read_csv("netflix_titles.csv")
```

## Dataset Description
The dataset contains information about Netflix titles such as:
- Type (Movie / TV Show)
- Title
- Director
- Cast
- Country
- Release year
- Date added
- Rating
- Duration
- Genre

## Data Cleaning
Missing values and incorrect formats can affect analysis.
So we handle missing values and convert columns into proper formats.


```python
df.isnull().sum()
```

```python
df['director'] = df['director'].fillna('Not Available')
df['cast'] = df['cast'].fillna('Not Available')
df['country'] = df['country'].fillna('Not Available')
df['rating'] = df['rating'].fillna('Not Available')
df['date_added'] = df['date_added'].fillna('Not Available')
df['duration'] = df['duration'].fillna('Not Available')
//This step ensures the dataset is consistent and ready for analysis.
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

## Feature Engineering
New features like year_added and month_added are created from the
date_added column to analyze year-wise and month-wise trends.


## Feature Engineering
Feature engineering is used to create new meaningful columns from
existing data. Here, year and month are extracted from date_added
to analyze year-wise trends.


```python
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df[['date_added', 'year_added', 'month_added']].head()

```

```python
df['year_added'].isnull().sum()
```

## Content Type Analysis
This section analyzes the distribution of Movies and TV Shows on Netflix.


```python
df['type'].value_counts()
```

```python
country_df = df['country'].str.split(', ').explode()
country_df.value_counts().head(10)
```

## Movies vs TV analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
//Netflix has more movies than shows
```

## Top 10 Genres

```python
genres = df['listed_in'].str.split(', ')
genres = genres.explode()

top_genres = genres.value_counts().head(10)

plt.figure()
top_genres.plot(kind='bar')
plt.title("Top 10 Genres on Netflix")
plt.ylabel("Count")
plt.show()

```

## Content added per year


```python
plt.figure()
df['year_added'].value_counts().sort_index().plot(kind='line')
plt.title("Content Added Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
** Netflix starts growing after 2015 **
```

```python
plt.figure()
df['year_added'].value_counts().sort_index().plot(kind='line')
plt.title("Content Added Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

** Netflix content incresed rapidly after 2015 **
